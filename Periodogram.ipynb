```python
import lightkurve as lk
%matplotlib inline
```

```python
lc = lk.search_lightcurve('KIC 10264202', author="Kepler", quarter=10).download()
lc.plot();
```

```python
pg = lc.normalize(unit='ppm').to_periodogram()
pg
```

```python
pg.plot();
```

```python
pg.plot(scale='log');
```

```python
period = pg.period_at_max_power #estimating period at which powe is maximum
period
```

```python
lc.fold(period).scatter(label=f'Period = {period.value:.3f} d'); # fold() function to see if it causes the eclipsing binary transits to align
```

```python
lc.fold(2*period).scatter(label=fr'Period = {2*period.value:.5f} d');
```
